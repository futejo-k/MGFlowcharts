## Collatz problem
[Collatz problem na Wikipedii](https://cs.wikipedia.org/wiki/Collatz%C5%AFv_probl%C3%A9m)

```python
i = 420

while i != 1:
    print(i)
    if i % 2 == 0:
        i /= 2
    else:
        i = 3*i + 1
print(i)
```
